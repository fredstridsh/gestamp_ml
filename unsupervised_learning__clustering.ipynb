# Introduction to Clustering

In this session, we explore **Unsupervised Learning**. It is the process of finding hidden patterns in data without pre-set labels. We will be testing two primary algorithms in the clustering world:

---

### KMeans: The Centroid-Based Specialist
* **Mechanism**: Groups data by finding the center points (centroids) of clusters and assigning points to the nearest center.
* **Best for**: Well-separated, circular "blobs" of data.
* **Requirement**: You must specify the number of groups ($k$) before running the algorithm.


### DBSCAN: The Density-Based Expert
* **Mechanism**: Groups data based on density, identifying areas where points are packed closely together.
* **Best for**: Complex shapes, curves, and non-linear patterns.
* **Outliers**: Unlike KMeans, it automatically detects and labels noise points that do not belong to a dense cluster.


---

### Workshop Objectives
We will use **Scikit-Learn** to:
1.  **Generate** synthetic datasets including Blobs and Moons.
2.  **Apply the Elbow Method** to determine the optimal cluster count for KMeans.
3.  **Visualize** results to understand where each algorithm succeeds or fails based on data geometry.

```python
# Start by importing the right packages
# We use matplotlib.pyplot for plotting, sklearn.datasets to generate
# blob-like and moon-like groups of data, sklearn.cluster for algorithms
# related to clustering, and numpy for calculations.

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons
from sklearn.cluster import KMeans, DBSCAN
import numpy as np

# Helper function to plot clusters. You do not need to go through this
def plot_clusters(X, labels, title):
    plt.figure(figsize=(8, 5))
    unique_labels = set(labels)
    colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))

    for k, col in zip(unique_labels, colors):
        if k == -1:
            col = 'k'

        class_member_mask = (labels == k)
        xy = X[class_member_mask]
        plt.scatter(xy[:, 0], xy[:, 1], c=[col], label=f'Cluster {k}' if k != -1 else 'Noise')

    plt.title(title)
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.legend()
    plt.show()
```

```python
X_blobs, _ = make_blobs(250, centers=3, random_state=42)
X_moons, _ = make_moons(250, noise=0.05, random_state=42)

# Elbow Method to find optimal K
inertia = []
k_range = range(2, 11)
for k in k_range:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_blobs)
    inertia.append(km.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, 'bx-')
plt.xlabel('k (Number of Clusters)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()
```

According to the results in the Elbow plot, what would be the optimal number of clusters?

```python
# We start with kmeans clustering, which is usually a quick, first attempt.
# To use kmeans clustering, we need to provide the code with a number of clusters
# it should look for. This is done in the parameter n_clusters. Swap 'XX' from
# the n_clusters=XX to the optimal number of clusters."
# random_state is there to make the code reproducable.

# Based on the elbow, we choose  XX
kmeans_blobs = KMeans(n_clusters=XX, random_state=42)
labels_blobs = kmeans_blobs.fit_predict(X_blobs)
plot_clusters(X_blobs, labels_blobs, "KMeans on Blobs")
```

```python
# Next, let us try DBSCAN. For data that is non-spherical, basing the clusters
# on density might be a better approach. Here he have two parameters, eps
# which defines the neighborhood distance for each point, and min_samples, the
# minimum neighbors needed for a point to be considered part of the neighborhood.
# can you find parameters XX, YY that yields robust results?

dbscan_blobs = DBSCAN(eps=XX, min_samples=YY)
labels_db_blobs = dbscan_blobs.fit_predict(X_blobs)
plot_clusters(X_blobs, labels_db_blobs, "DBSCAN on Blobs")
```

```python
# KMeans on Moons (Trying to find XX clusters)
kmeans_moons = KMeans(n_clusters=XX, random_state=42)
labels_km_moons = kmeans_moons.fit_predict(X_moons)
plot_clusters(X_moons, labels_km_moons, "KMeans on Moons (Struggles with Geometry)")
```

```python
# DBSCAN on Moons (Better for non-linear shapes)
# Scikit-learn's 'min_samples' is equivalent to Sealion's 'min_neighbors'
dbscan_moons = DBSCAN(eps=XX, min_samples=YY)
labels_db_moons = dbscan_moons.fit_predict(X_moons)
plot_clusters(X_moons, labels_db_moons, "DBSCAN on Moons (Handles Geometry Well)")
```
